```python
dfcovid.drop(columns = ['Island','WHO Western Pacific Region','Female head of government',
                        'All-cause death count, 2017','All-cause death count, 2018',
                        'All-cause death count, 2019','All-cause death count, 2020',
'Excess deaths in 2020 per 100,000 population, relative to 2017-2019 average'], inplace = True)
```

```python
dfcov = dfcovid.dropna()
```

```python
dfcov.isnull().sum()
```

```python
df21cov = pd.merge(df21,dfcov, how = 'left', left_on = 'Country name', right_on = 'Country name')
```

```python
df21cov.info()
```

```python
# Change comma into dot
df21cov['COVID-19 deaths per 100,000 population in 2020'] = df21cov['COVID-19 deaths per 100,000 population in 2020'].str.replace(',','.')
df21cov['Median age'] = df21cov['Median age'].str.replace(',','.')
df21cov['Index of exposure to COVID-19  infections in other countries as of March 31'] = df21cov['Index of exposure to COVID-19  infections in other countries as of March 31'].str.replace(',','.')
df21cov['Log of average distance to SARS countries'] = df21cov['Log of average distance to SARS countries'].str.replace(',','.')
df21cov['Index of institutional trust'] = df21cov['Index of institutional trust'].str.replace(',','.')
df21cov['Gini coefficient of income'] = df21cov['Gini coefficient of income'].str.replace(',','.')

# convert from string into numeric
df21cov['COVID-19 deaths per 100,000 population in 2020'] = pd.to_numeric(df21cov['COVID-19 deaths per 100,000 population in 2020'])
df21cov['Median age'] = pd.to_numeric(df21cov['Median age'])
df21cov['Index of exposure to COVID-19  infections in other countries as of March 31'] = pd.to_numeric(df21cov['Index of exposure to COVID-19  infections in other countries as of March 31'])
df21cov['Log of average distance to SARS countries'] = pd.to_numeric(df21cov['Log of average distance to SARS countries'])
df21cov['Index of institutional trust'] = pd.to_numeric(df21cov['Index of institutional trust'])
df21cov['Gini coefficient of income'] = pd.to_numeric(df21cov['Gini coefficient of income'])
```

```python
df21cov.drop(columns = ['Explained by: Log GDP per capita','Explained by: Social support','Explained by: Healthy life expectancy','Explained by: Freedom to make life choices','Explained by: Generosity','Explained by: Perceptions of corruption','Dystopia + residual', 'Population 2020', 'Population 2019','Standard error of ladder score', 'upperwhisker', 'lowerwhisker','Ladder score in Dystopia'], inplace = True)
```

```python
df21cov.isnull().sum()
```

```python
df21cov= df21cov.dropna()
```

# Exploratory Data Analysis (EDA) 

```python
df21cov
```

```python
plt.figure(figsize = (12,12))
plt.figure(figsize=(8, 12))
sns.heatmap(df21cov.corr()[['Ladder score']].sort_values('Ladder score', ascending= False),
            annot=True,
            cmap='coolwarm',
            vmin=-1, 
            vmax=1);
```

```python
plt.figure(figsize = (12,12))
plt.scatter(df21cov['COVID-19 deaths per 100,000 population in 2020'], df21cov['Ladder score'], s = 20)
plt.xlabel('COVID-19 deaths per 100,000 population in 2020')
plt.ylabel('Happiness')
plt.title('Happiness vs Covid-19 deaths', fontsize = 20)
;
```

```python
plt.figure(figsize = (12,8))
sns.barplot(x = 'COVID-19 deaths per 100,000 population in 2020', y = 'Ladder score', data= df21cov.reset_index().sort_values('Ladder score'))
#plt.xticks([0,1,2],['less 5000 employees','5001 to 10000 employees','3 = 10000+ employees'],rotation = 20)
plt.title('Happiness vs Covid-19 deaths', fontsize = 20)
;
```

```python
plt.figure(figsize = (12,8))
sns.catplot(x = 'COVID-19 deaths per 100,000 population in 2020', y = 'Ladder score', data= df21cov.reset_index().sort_values('Ladder score'))
#plt.xticks([0,1,2],['less 5000 employees','5001 to 10000 employees','3 = 10000+ employees'],rotation = 20)
plt.title('Happiness vs Covid-19 deaths', fontsize = 20)
;
```
